```python
import pandas as pd 
import json 
import os
```

```python
# converting each annotation category into dataframe 

# Amusement
amusement = pd.DataFrame()
amusement_files = os.listdir("./annotation/amusement")
for file in amusement_files:
    file_name = "./annotation/amusement/" + file
    pd_from_json = pd.DataFrame()
    with open(file_name) as f:
        data = json.load(f)
        pd_from_json = pd.json_normalize(data)
    # concatenate to existing dataframe 
    amusement = pd.concat([amusement, pd_from_json])
f.close()
print(f"Number of annotations for amusement = {len(amusement)}")

# Anger 
anger = pd.DataFrame()
anger_files = os.listdir("./annotation/anger")
for file in anger_files:
    file_name = "./annotation/anger/" + file
    pd_from_json = pd.DataFrame()
    with open(file_name) as f:
        data = json.load(f)
        pd_from_json = pd.json_normalize(data)
    # concatenate to existing dataframe 
    anger = pd.concat([anger, pd_from_json])
f.close()
print(f"Number of annotations for anger = {len(anger)}")

# Awe 
awe = pd.DataFrame()
awe_files = os.listdir("./annotation/awe")
for file in awe_files:
    file_name = "./annotation/awe/" + file
    pd_from_json = pd.DataFrame()
    with open(file_name) as f:
        data = json.load(f)
        pd_from_json = pd.json_normalize(data)
    # concatenate to existing dataframe 
    awe = pd.concat([awe, pd_from_json])
f.close()
print(f"Number of annotations for awe = {len(awe)}")

# Contentment
contentment = pd.DataFrame()
contentment_files = os.listdir("./annotation/contentment")
for file in contentment_files:
    file_name = "./annotation/contentment/" + file
    pd_from_json = pd.DataFrame()
    with open(file_name) as f:
        data = json.load(f)
        pd_from_json = pd.json_normalize(data)
    # concatenate to existing dataframe 
    contentment = pd.concat([contentment, pd_from_json])
f.close()
print(f"Number of annotations for contentment = {len(contentment)}")

# Disgust
disgust = pd.DataFrame()
disgust_files = os.listdir("./annotation/disgust")
for file in disgust_files:
    file_name = "./annotation/disgust/" + file
    pd_from_json = pd.DataFrame()
    with open(file_name) as f:
        data = json.load(f)
        pd_from_json = pd.json_normalize(data)
    # concatenate to existing dataframe 
    disgust = pd.concat([disgust, pd_from_json])
f.close()
print(f"Number of annotations for disgust = {len(disgust)}")

# Excitement
excitement = pd.DataFrame()
excitement_files = os.listdir("./annotation/excitement")
for file in excitement_files:
    file_name = "./annotation/excitement/" + file
    pd_from_json = pd.DataFrame()
    with open(file_name) as f:
        data = json.load(f)
        pd_from_json = pd.json_normalize(data)
    # concatenate to existing dataframe 
    excitement = pd.concat([excitement, pd_from_json])
f.close()
print(f"Number of annotations for excitement = {len(excitement)}")

# Fear 
fear = pd.DataFrame()
fear_files = os.listdir("./annotation/fear")
for file in fear_files:
    file_name = "./annotation/fear/" + file
    pd_from_json = pd.DataFrame()
    with open(file_name) as f:
        data = json.load(f)
        pd_from_json = pd.json_normalize(data)
    # concatenate to existing dataframe 
    fear = pd.concat([fear, pd_from_json])
f.close()
print(f"Number of annotations for fear = {len(fear)}")

# Sadness
sadness = pd.DataFrame()
sadness_files = os.listdir("./annotation/sadness")
for file in sadness_files:
    file_name = "./annotation/sadness/" + file
    pd_from_json = pd.DataFrame()
    with open(file_name) as f:
        data = json.load(f)
        pd_from_json = pd.json_normalize(data)
    # concatenate to existing dataframe 
    sadness = pd.concat([sadness, pd_from_json])
f.close()
print(f"Number of annotations for sadness = {len(sadness)}")
```

```python
# combine everything together 
annotations = pd.concat([amusement, anger, awe, contentment, disgust, excitement, fear, sadness])

print(f"Total annotations = {len(annotations)}")
print(annotations.head())
```
